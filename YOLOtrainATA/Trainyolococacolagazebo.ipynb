```python
import os
from ultralytics import YOLO

#load model
model=YOLO('yolov8n.yaml')
model.to('cuda')

#Use the model

results=model.train(data=os.path.join(ROOT_DIR,"coke.yaml"),epochs=100, project=os.path.join(ROOT_DIR,"results_coke2"))

```

```python
import os
from ultralytics import YOLO
model=YOLO(os.path.join(ROOT_DIR,"results_coke2/train5/weights/best.pt"))
results=model.track(source=os.path.join(ROOT_DIR,"predict/Final.mp4"), show=True, save = True, project = '/content/drive/MyDrive/YOLOtrainATA/results')
```
